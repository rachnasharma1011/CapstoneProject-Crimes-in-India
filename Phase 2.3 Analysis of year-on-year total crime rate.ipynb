```python
columns = data.columns[1:]
```

```python
for i in columns:
    plt.figure(figsize=(10, 5))
    sns.barplot(x='States',y=i, data=data)
    plt.xticks(rotation=90)
    plt.show()
```

```python
#From the above bagraphs, it is observed that:
#Year 2018 & 2019
#1. Crime rate was highest in Delhi followed by Kerala and Andaman and Nicobar Islands.
#2. Crime rate was minimum in Ladakh and Daman & Diu
#3. Although crime rate has reduced in keral in year 2019 as compared to year 2018.
#For year 2020
#Tamil Nadu has shown the highest crime rate in the year 2020
# There was a spike in crime rate in the state of Kerala and Gujarat and crimes were also observed in Daman & Diu and Andaman and Nikobar Islands.
#Year 2021
#Crime rate in Kerala and Delhi has shown a sharp increase in crime rate while there is a decrease in Delhi's crime rate 
```

```python
df=pd.DataFrame({'State':data['States'], 'Crime Rate 2018':data['Crime Rate 2018'], 'Crime Rate 2019':data['Crime Rate 2019'], 'Crime Rate 2020':data['Crime Rate 2020'], 'Crime Rate 2021':data['Crime Rate 2021']})
df_sorted_2018=df.sort_values('Crime Rate 2018', ascending=False)
df_sorted_2019=df.sort_values('Crime Rate 2019', ascending=False)
df_sorted_2020=df.sort_values('Crime Rate 2020', ascending=False)
df_sorted_2021=df.sort_values('Crime Rate 2021', ascending=False)      
df_sorted_2018.head()                                                                                                                                                
```

```python
df_sorted_2018.tail()
```

```python
plt.figure(figsize=(10,5))
plt.bar('State', 'Crime Rate 2018', data=df_sorted_2018, color='green')
plt.xlabel("State", size=15)
plt.ylabel("Crime Rate 2018", size=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#1. Crime rate was highest in Delhi followed by Kerala, Andaman and Nicobar Islands, Haryana and Tamil Nadu.
#2. Crime rate was nil in Ladakh and Daman & Diu
```

```python
plt.figure(figsize=(10,5))
plt.bar('State', 'Crime Rate 2019', data=df_sorted_2019, color='red')
plt.xlabel("State", size=15)
plt.ylabel("Crime Rate 2019", size=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#1. Crime rate was highest in Delhi followed by Kerala, Andaman and Nicobar Islands, Haryana and Tamil Nadu.
#2. Crime rate was nil in Ladakh and Daman & Diu
#3. Although crime rate has reduced in Kerala in year 2019 as compared to year 2018.
```

```python
plt.figure(figsize=(10,5))
plt.bar('State', 'Crime Rate 2020', data=df_sorted_2020, color='brown')
plt.xlabel("State", size=15)
plt.ylabel("Crime Rate 2020", size=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#There was a increase in crime rate in Tamil Nadu, it has shown the highest crime rate in the year 2020
# There was a spike in crime rate in the state of Kerala and Gujarat.
# Crimes were also observed in Daman & Diu and Andaman and Nikobar Islands.
```

```python
plt.figure(figsize=(10,5))
plt.bar('State', 'Crime Rate 2021', data=df_sorted_2021, color='blue')
plt.xlabel("State", size=15)
plt.ylabel("Crime Rate 2021", size=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# Crime rate in Kerala and Delhi has shown a sharp increase and they are showing the highest crime rate. While there is a decrease in Tamil Nadu's crime rate
```
